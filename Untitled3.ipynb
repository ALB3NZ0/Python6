Цель работы: На данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков.

```python
import matplotlib.pyplot as plt
import pandas as pd

starwars = pd.read_csv("star_wars_character_dataset.csv")
starwars
```

```python


plt.plot(starwars['name'].head(5), starwars["mass"].head(5), color="pink", label="Продолжительность")
plt.xlabel('Имя')
plt.ylabel("Масса")
plt.title('Масса персонажа')
plt.grid()
plt.show()
```

Данный линейный график нам показывает, что самая большая масса у Darth Vаder составляет 135 кг, а самый легкий это R2-D2, его вес 30 кг.

```python
plt.figure(figsize=(20, 6))
plt.plot(starwars['name'].head(10), starwars["height"].head(10), color="blue", label="Рост")
plt.xlabel ('Персонажи')
plt.ylabel("Рост персонажа")
plt.title('Рост')
plt.grid()
plt.show()
```

Данный линейный график нам показывает, что самый высокий это Darth Vаder его рост составляет +-210 см , а самый нзикий это R2-D2 и R5-D4, их рост составляет примерно +- 90 см.

```python
plt.plot(starwars['name'].head(5), starwars["eye_color"].head(5), color="purple", label="Тип")
plt.xlabel ('Персонажи')
plt.ylabel("Цвета глаз")
plt.title('Тип глаз ')
plt.grid()
plt.show()
```

Этот линейный график показывает нам цвет глаз персонгаж,вот у Люка Скайволкер синий цвет галз, а у Дарт Вейдера желтый цвет галз.

```python
plt.figure(figsize=(20, 6))
plt.bar(starwars['name'].head(10), starwars["sex"].head(10), color="black", label="пол")
plt.xlabel('Имя персонажей')
plt.ylabel('Виды пол')
plt.title('Пол')
plt.legend()
plt.show()
```

Далее был создан столбчатый график который показывает, Пол персонажей. Darth Varder имеет мужиской пол , а Leia Organa имеет женский пол, а у R2-D2 нет  своего пола.

```python
plt.figure(figsize=(20, 6))
plt.bar(starwars['name'].head(10), starwars["gender"].head(10), color="green", label="Гендер")
plt.xlabel('Имя персонажей')
plt.ylabel('Гендеры')
plt.title('Гендер')
plt.legend()
plt.show()
```

Далее был создан столбчатый график который показывает Гендер персонажей. Leia Organa женский гендер,а вот Darth Varder имеет мужской гендер также и R2-D2 тоже имеет мужской гендер.  

```python
plt.figure(figsize=(20, 6))
plt.bar(starwars['name'].head(10), starwars["species"].head(10), color="yellow", label="Виды персонажей")
plt.xlabel('Имя персонажей')
plt.ylabel('Вид')
plt.title('Виды персонажей')
plt.legend()
plt.show()
```

Далее был опять создан столбчатый график, который показывает вид персонажей. Obi-Wan Kenobi человек, а вот R2-D2 робот.

```python
data = starwars["homeworld"].value_counts().head()
plt.pie(data, labels = data.index, colors = ["blue", "red","pink","orange","purple"],autopct="%1.1f%%")
plt.title("Соотношение персонажей откуда они родом")
plt.legend()
plt.show()
```

Далее был создана круговая диаграмма. Она показывает, что самая густо населенная страна персонажей это Naboo, а самая не популярная Coruscant,Alderaan,Kamino.




```python
data = starwars["vehicles"].value_counts().head()
plt.figure(figsize=(6,6))
plt.pie(data, labels = data.index, colors = ["blue", "red","pink","orange","purple"],autopct="%1.1f%%")
plt.title("Траснпортые средства")
plt.legend()
plt.show()
```

Далее был создана круговая диаграмма. Она показывает, что самый популярный транспорт является Tribubble bongo, а меньше всего используют AT-ST, Zephyr-G swoop bike, XJ-6 airspeeder, Imperial Speeder, Snowspeeder,Imperial Speeder Bike.

```python
plt.scatter(starwars["birth_year"].head(20), starwars["height"].head(20))
plt. title( 'Связь между День рождением и ростом' )
plt.xlabel('День рождения')
plt.ylabel('Рост персонажа')
plt.show()
```

Данная диаграмма рассеивания показывает отсуствие связи между массой и ростом.

```python
plt.hist(starwars["eye_color"].head(20), bins=30, edgecolor="black")
plt.xlabel('Цвет')
plt.ylabel('Количество')
plt.title('Цвета глаз')
plt.legend()
plt.show()
```

Данная гистограмма показывает нам цвет глаз. Самый популярный это Blue, а самый не популярный это blue-gray, black,orange,hazel.

```python
plt.hist(starwars["mass"].head(20), bins=30, edgecolor="black")
plt.xlabel('Масса')
plt.ylabel('Количество')
plt.title('Вес персонажей')
plt.legend()
plt.show()
```

Данная гистограмма показывает нам вес персонажей. Самый популярный это +-100кг, а самый не популярный это +-1300кг.
